По цьому графіку бачимо більш менш очікуване, 
з цікавого те, що другу найбільшу зарплату отримують люди що зараз у неоплачуваній відпустні, а прямо за ними йдуть ті хто тимчасово не працюють. 
Найменшу ж зарплатню отримують люди що працюють парт-тайм 


```python
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = pd.get_dummies(df[categorical_cols], drop_first=True)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.drop('Зарплата / дохід в IT у $$$ за місяць, лише ставка \nЧИСТИМИ - після сплати податків')
df_numeric = df[numerical_cols]
df_for_corr = pd.concat([df_numeric, df_encoded], axis=1)
corr_with_salary = df_for_corr.corrwith(df['Зарплата / дохід в IT у $$$ за місяць, лише ставка \nЧИСТИМИ - після сплати податків'])

corr_sorted = corr_with_salary.sort_values(ascending=False)
print("Найбільші позитивні кореляції з зарплатою:")
print(corr_sorted.head(10))
print("\nНайбільші негативні кореляції з зарплатою:")
print(corr_sorted.tail(10))
```

```python
top_pos = corr_sorted.head(10)
top_neg = corr_sorted.tail(10)
top_corr = pd.concat([top_pos, top_neg])

heatmap_data = top_corr.to_frame(name='Кореляція із зарплатою')
plt.figure(figsize=(8, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0, cbar_kws={'label': 'Кореляція'})
plt.title('Heatmap топ 20 кореляцій із зарплатою')
plt.ylabel('Фактори')
plt.tight_layout()
plt.show()
```

Далі я хотів подивитися на вплив факторів на зростання заробітної плати, 
Побудував таблицю кореляцій і представив 10 найбільших. 
Але в даному випадку “найбільших” є дуже гучним словом, бо єдиною значущою кореляцію тут є ваш досвід, 
за ним з більш менш гарним значенням йде займання рівня сеньор, ну а усі інші чинники мають невиликий позитивний вплив. 
Після цього мені стало дуже цікаво подивитися які змінні мають негативний вплив, тому саме це я і зробив, знову таки, 
їх коефіцієнти не є значущими, проте все ж таки значення мають, 
поперше бачимо що найбільший негативний коефіцієнт займає позиція Джуніор, 
за ним Міддл, а ось інтерн має більш привабливий коефіцієнт, 
проте це скоріш за все через те що записів у таблиці з інтернами буле дуже мало, і тому цьому довіряти не слід.

```python
median_salary_by_exp = df.groupby('Загальний стаж роботи за нинішньою ІТ-спеціальністю')[salary_col ].median().reset_index()

plt.figure(figsize=(10,6))
sns.scatterplot(data=median_salary_by_exp, x='Загальний стаж роботи за нинішньою ІТ-спеціальністю', y=salary_col , s=100)
z = np.polyfit(median_salary_by_exp['Загальний стаж роботи за нинішньою ІТ-спеціальністю'], median_salary_by_exp[salary_col ], 1)
p = np.poly1d(z)
plt.plot(median_salary_by_exp['Загальний стаж роботи за нинішньою ІТ-спеціальністю'], p(median_salary_by_exp['Загальний стаж роботи за нинішньою ІТ-спеціальністю']), color='red', linestyle='--', label='Лінійний тренд')
for i, row in median_salary_by_exp.iterrows():
    plt.text(row['Загальний стаж роботи за нинішньою ІТ-спеціальністю'], row[salary_col ] + 50,
             f"${int(row[salary_col ]):,}", 
             ha='center', va='bottom', fontsize=9, color='black')
plt.xlabel('Стаж роботи')
plt.ylabel('Медіанна зарплата, $')
plt.title('Зарплата залежно від стажу роботи')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_ua = df[df['Де ви зараз живете?'].str.lower().str.contains('в україні')]
df_foreign = df[df['Де ви зараз живете?'].str.lower().str.contains('за кордоном')]
ua_median_salary = df_ua[salary_col].median()
foreign_median_salary = df_foreign[salary_col].median()
ua_count = len(df_ua)
foreign_count = len(df_foreign)

plt.figure(figsize=(8, 5))
plt.bar(['Україна', 'За кордоном'], [ua_count, foreign_count], color=['red', 'skyblue'])
plt.title('Кількість респондентів')
plt.ylabel('Кількість осіб')
plt.text(0, ua_count + ua_count * 0.02, str(ua_count), ha='center', fontsize=12)
plt.text(1, foreign_count + foreign_count * 0.02, str(foreign_count), ha='center', fontsize=12)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(['Україна', 'За кордоном'], [ua_median_salary, foreign_median_salary], color=['skyblue', 'red'])
plt.title('Медіанна зарплата')
plt.ylabel('Зарплата, $')
plt.text(0, ua_median_salary + ua_median_salary * 0.02, f"${ua_median_salary:,.0f}", ha='center', fontsize=12)
plt.text(1, foreign_median_salary + foreign_median_salary * 0.02, f"${foreign_median_salary:,.0f}", ha='center', fontsize=12)
plt.show()
```

Зарплати в таких напрямках як Back-end розробка, Embedded, Front-end розробка та Full-stack розробка тримаються на однаковому рівні — близько $1000, що свідчить про приблизно однакову оцінку цих ролей для новачків.

Спеціалізації System development та Mobile розробка мають дещо нижчі медіанні зарплати ($950 та $825 відповідно), можливо через меншу кількість проектів або менші вимоги до початкових позицій.

```python
avg_experience_by_position = df.groupby('Оберіть ваш тайтл або роль у компанії')['Загальний стаж роботи за нинішньою ІТ-спеціальністю'].mean().sort_values(ascending=True)

plt.figure(figsize=(12, 8))
plt.plot(avg_experience_by_position.index, avg_experience_by_position.values, marker='o', linestyle='-', color='skyblue')
plt.title('Середній стаж роботи за позиціями')
plt.ylabel('Середній стаж роботи (роки)')
plt.xticks(rotation=45, ha='right')
for x, y in zip(avg_experience_by_position.index, avg_experience_by_position.values):
    plt.text(x, y + 0.1, f"{y:.1f}", ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

Очікувані стрибки у стажі між рівнями:
Intern → Junior: зростання приблизно на 1 рік, що логічно — більшість отримують позицію Junior після 6–12 місяців стажування/початкової роботи.
Junior → Middle: приріст на 2.5 роки, тобто зазвичай потрібно 3–4 роки загального досвіду, щоб стати впевненим мідлом.
Middle → Senior: ще пориблизно 3.5 роки. Відображає складність переходу, який вимагає глибшого технічного розуміння, самостійності, часто — менторства.
Senior → Tech Lead / Architect / Principal: тут діапазон досвіду розширюється, і ріст менш лінійний — усе залежить від компанії, навичок лідерства, спеціалізації.
Цікаві спостереження:
CEO / C-level ≈ Lead/team ≈ Senior — цікаво, що топ-менеджмент і технічні лідери мають схожий досвід (приблизно 7.3 року). Це може бути результатом того, що в ІТ стартапах С-level позиції часто займають досить "молоді" фахівці.
Tech Lead > Architect > Staff — хоча ці ролі іноді перетинаються, графік демонструє, що Tech Lead вимагає найбільше досвіду серед усіх, крім Principal.
Head має нижчий середній стаж (6.45) порівняно з Senior/Architect, що може свідчити про менеджерську роль, де ключову роль відіграють soft skills та організаційні здібності.
Висновок:
Шлях від Intern до Senior займає в середньому близько 7.5 років.
Позиції на кшталт Tech Lead, Architect, Principal вимагають 8+ років стабільного розвитку.
Є деякі ролі (Head, CEO), які можуть бути досягнуті швидше за рахунок нетехнічних навичок, підприємницької активності або кар’єрного повороту.
